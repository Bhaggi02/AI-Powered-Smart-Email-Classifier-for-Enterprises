```python
!pip install nltk pandas scikit-learn
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

```

```python
df = pd.read_csv('/content/final_data.csv')
df.head()
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_email(text):
    # Remove signatures
    text = re.sub(r"(regards|thanks|sincerely).*", "", text, flags=re.I)

    # Lowercase
    text = text.lower()

    # Remove email addresses & URLs
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'http\S+', '', text)

    # Remove special characters & numbers
    text = re.sub(r'[^a-z\s]', '', text)

    # Tokenization & stopword removal
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    return " ".join(words)
```

```python
df["clean_email"] = df["mail"].apply(preprocess_email)
df
```

```python
def assign_category(text):
    if "urgent" in text or "account" in text:
        return "Security"
    elif "meeting" in text or "report" in text:
        return "Work"
    elif "order" in text:
        return "Promotion"
    elif "reminder" in text:
        return "Reminder"
    else:
        return "Spam"

df["category"] = df["clean_email"].apply(assign_category)
```

```python
def assign_urgency(text):
    if "urgent" in text or "immediately" in text:
        return "High"
    elif "tomorrow" in text or "scheduled" in text:
        return "Medium"
    else:
        return "Low"

df["urgency"] = df["clean_email"].apply(assign_urgency)
```
